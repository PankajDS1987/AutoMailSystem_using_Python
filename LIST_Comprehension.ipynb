```python
# FLATTEN the LIST
def flatten_list (a):
  updated_list=[]

  while a:
    ele = a.pop()
    if type(ele) == list:
      a.extend(ele)
    else:
      updated_list.append(ele)
  updated_list.sort()
  return updated_list
```

```python
a = [4,[3,5,5,[4, 6,6 ],[4, [5, 6,[9]]]]]
#a = [2,3,[4,5]]
xx = flatten_list (a)
print(xx)
```
